```python
import time
import random
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def complex_time_list(n):
    arry1 = [random.random() for i in range(50000)] # a list of random 'data'
    t=[]
    d_keys = list(range(n)) 
    
    # calculate time period for deleting the first element on the list:
    start =  time.perf_counter()
    for k in d_keys:     
        del arry1[0]
    end =  time.perf_counter()
    t.append(end-start)
    return t
```

```python
def complex_time_dic(n):
    t=[]
    arry1 = [random.random() for i in range(50000)]
    d1 = dict(enumerate(arry1)) # Create a simple dictionary of key:val pairs using these two lists
    d_keys = list(d1.keys()) 
    
      # calculate time period for deleting the k+1th element on the dictionary
    start = time.perf_counter()
    for k in d_keys:
        del d1[k] 
    end = time.perf_counter()
    t.append(end-start)
    return t
```

## Testing

```python
list_time=[]
for r in range(1000,1500):
    list_time += complex_time_list(r)   

plt.plot(list_time)
plt.ylim
plt.show()
```

```python
dic_time=[]
for r in range(1000,1500):
    dic_time += complex_time_dic(r)
    
plt.plot(dic_time)
plt.ylim
plt.show()
```
